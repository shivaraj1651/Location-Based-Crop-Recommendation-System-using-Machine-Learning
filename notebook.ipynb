```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler

from sklearn.svm import SVC
# import pickle

import joblib
```

```python
dataset=pd.read_csv('Crop_recommendation.csv',header=None)
x=dataset.iloc[:,[0,3,4,5]].values
print(x)
```

```python
y=dataset.iloc[:,[7]].values
print(y)
```

```python
# from sklearn.preprocessing import LabelEncoder, StandardScaler
# label_encoder = LabelEncoder()
# y = label_encoder.fit_transform(y)

y
```

```python
xtest,xtrain,ytest,ytrain=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
classifier=KNeighborsClassifier(n_neighbors=3,weights = 'uniform',algorithm = 'brute')
ytrain=np.ravel(ytrain)
classifier.fit(xtrain,ytrain)
# model=open("newknnmodel.pkl",'wb')
# pickle.dump(classifier,model)
# model.close()

joblib.dump(classifier,"newknnmodel.pkl")
```

```python
from sklearn.metrics import accuracy_score
```

```python
ypred=classifier.predict(xtest)
score=accuracy_score(ytest,ypred)
print(f"Accuracy score of KNN = {score}")
```

```python
classifier=tree.DecisionTreeClassifier(max_depth=3,min_samples_leaf=3)
ytrain=np.ravel(ytrain)
classifier.fit(xtrain,ytrain)
# model1=open("newdtmodel.pkl",'wb')
# pickle.dump(classifier,model1)
# model1.close()

joblib.dump(classifier,"newdtmodel.pkl")
```

```python
ypred=classifier.predict(xtest)
score=accuracy_score(ypred,ytest)
print(f"Accuracy score of Decision Tree = {score}")
```

```python
classifier=SVC(kernel='rbf', random_state = 100)
ytrain=np.ravel(ytrain)
classifier.fit(xtrain,ytrain)

# model2=open("newsvmmodel.pkl",'wb')
# pickle.dump(classifier,model2)
# model2.close()

joblib.dump(classifier,"newsvmmodel.pkl")
```

```python
ypred=classifier.predict(xtest)
score=accuracy_score(ypred,ytest)
print(f"Accuracy score of SVM = {score}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
classifier = RandomForestClassifier(n_estimators=100, random_state=42)

ytrain=np.ravel(ytrain)
classifier.fit(xtrain,ytrain)
# model2=open("randommodel.pkl",'wb')
# pickle.dump(classifier,model2)
# model2.close()

joblib.dump(classifier,"randommodel.pkl")
```

```python
ypred=classifier.predict(xtest)
score=accuracy_score(ytest,ypred)
print(f"Accuracy score of RF = {score}")
```
