```python
import pandas as pd
```

```python
x=dataset.iloc[:,[0,1,2,3,4,5,6]].values

```

```python
x[0]
```

```python
y=dataset.iloc[:,[7]].values
```

```python
y[0]
```

```python
print(len(x),len(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
#ML (Algorithm) we are going to use is KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier()
# this line creates an object of the class KNN
```

```python
#let us traing the model using training data
import numpy as np
ytrain=np.ravel(ytrain)
model.fit(xtrain,ytrain)
```

```python
#let us test our model
testdata=np.array([[69,55,38,22.70883798,82.63941394,5.70080568,271.3248604]])
result=model.predict(testdata)
print(f"Resultof prediction ={result[0]}")
```

```python
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accu=accuracy_score(ypred,ytest)
print(f"accuracy of the model = {accu*100}")
```
